```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
from sklearn import tree


clf = tree.DecisionTreeClassifier(random_state=0)
iris = load_iris()

clf = clf.fit(iris.data, iris.target)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```

```python
print(tree.export_text(clf))
```

### If we are drawing a decision tree it will be constructed till its depth, and this may lead to OVERFITTING. To overcome this situation sklearn provide some parameters like MINI_SAMPLE LEAF, MAX_DEPTH. Different values of max_depth can be used and the accuracy is checked. 
### After the construction of decision tree we can PRUNE(cutting) the TREE so that it maynot lead to overfitting. This is called as COST COMPLEXITY PRUNING. 
#### WE provide CCP value which will restrict the decision tree to some depth.

```python
print(__doc__)
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
```

#### here i am using brest cancer data which has already been present in sklearn. we just have to import it here. It has two outputs- MALGNANRT, BENIGN.

```python
X, y = load_breast_cancer(return_X_y=True)

#Splitting the datset into train adn test data.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

#Initialization of classifer
clf = DecisionTreeClassifier(random_state=0)

#fitting classifier on X_train and y_train
clf.fit(X_train,y_train)

# the default criterai used here is GINI. We can use entropy for INFORMATION GAIN.
#one que is-- WHY GINI?
# If you look at the graph of info gain & gini it will be binomial, but the height of the GINI graph is half than info gain. 
#This leads to reduction in computation time by half.
```

```python
pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```
